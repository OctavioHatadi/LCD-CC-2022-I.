###Multiprocessing

Es un paquete en python que permite la creación de procesos y ofrece concurrencia local. 

Una manera censilla de crear un proceso es por medio de la construcción de un objeto de tipo 'Process' e invocarlo por medio del método 'start()'

```python
import multiprocessing as mp
def tarea(cadena): 
  print('Hola', cadena) 

if __name__== '__main__':    #Permite diferencias si lo estamos ejecutando como el programa principal o no
  p = mp.Process(target=tarea, args=('Lucas C.',)) #Se crea un objeto de tipo 'proces', se le pide que haga una tarea 
  #La coma permite decirle que espere mas argumentos 
  p.start() #Iniciamos el proceso
  p.join() #Regresar la ruta de ejecución al padre
```

```python
#EJEMPLO 2
import time

def cacl_cuad(numeros): 
  print('Calcula el cuadrado de los numeros')
  for n in numeros:
    time.sleep(0.2)
    print('Cuadrado:', n*n)

nums= range(10)
t=time.time()
pl = mp.Process(target=cacl_cuad, args=(nums, ))
pl.start()
pl.join()

print( "Tiempo de proceso: ", time.time()-t)
print ('Termina la ejecución númerica')
```
